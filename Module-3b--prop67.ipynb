```python
import pandas as pd
pd.__version__
```

```python
# Chapter 5
props = pd.read_csv('../data/committees.csv')
props.head()
```

```python
contribs = pd.read_csv('../data/contributions.csv')
contribs.head()
```

```python
props.prop_name.value_counts()
```

```python
prop_67 = props[props.prop_name == 'PROPOSITION 067- REFERENDUM TO OVERTURN BAN ON SINGLE-USE PLASTIC BAGS.']
prop_67
```

```python
prop_67.info()
```

```python
contribs.info()
```

```python
# merge the contributions and propostions dataset relevant to prop_67
merged = pd.merge(prop_67, contribs, on='calaccess_committee_id')
merged.head()
```

```python
merged.info()
```

```python
# total $ spent on campaigns pro and con
merged.amount.sum()
```

```python
# Which side got more large donations?
merged.committee_position.value_counts()
```

```python
support = merged[merged.committee_position == 'SUPPORT']
support.amount.sum()
```

```python
oppose = merged[merged.committee_position == 'OPPOSE']
oppose.amount.sum()
```

```python
support_vs_oppose = support.amount.sum() / merged.amount.sum()
support_vs_oppose
```

```python
# what were the ten biggest contributors?
top_10_support = support.sort_values('amount', ascending=False)
top_10_support.head(10)
```

```python
top_10_oppose = oppose.sort_values('amount', ascending=False)
top_10_oppose.head(10)
```

```python
# group the two committee_name_ columns as they are just duplicates due to previous merge
merged.groupby('committee_name_x').amount.sum().reset_index().sort_values('amount', ascending=False)
```

```python
# Finding the top contributors
merged.groupby(["contributor_firstname", "contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```

## Extra questions to answer:

1. What percentage of donations came from people who live outside of California?
2. What are the top employers of donors who gave for and against the measure?
3. Which committees had the fewest donors?

```python
# Q1 - What percentage of donations came from people who live outside of California?
outside = merged[merged.contributor_state != 'CA']
inside = merged[merged.contributor_state == 'CA']

total = merged.amount.sum()
o = outside.amount.sum()
i = inside.amount.sum()

pct = total / 100
[o / pct, i / pct]
```

__Answer #1__

Donations from Californian donors (34.23%) were largely outweight by non Californians (65.76%) 2 to 1

```python
# Q2 - What are the top employers of donors who gave for and against the measure?

oppose_employers = oppose.contributor_employer.value_counts()
support_employers = support.contributor_employer.value_counts()

[oppose_employers.head(), support_employers.head()]
```
